```python
# COLUMN: Page Views

# identify the quartiles
q1, q3 = np.percentile(df['Page Views'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Page Views'] >= lower_bound) & (df['Page Views'] <= upper_bound)]

# COLUMN: Session Duration

# identify the quartiles
q1, q3 = np.percentile(df['Session Duration'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Session Duration'] >= lower_bound) & (df['Session Duration'] <= upper_bound)]


# COLUMN: Bounce Rate

# identify the quartiles
q1, q3 = np.percentile(df['Bounce Rate'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Bounce Rate'] >= lower_bound) & (df['Bounce Rate'] <= upper_bound)]

# COLUMN: Time on Page

# identify the quartiles
q1, q3 = np.percentile(df['Time on Page'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Time on Page'] >= lower_bound) & (df['Time on Page'] <= upper_bound)]

# COLUMN: Previous Visits

# identify the quartiles
q1, q3 = np.percentile(df['Previous Visits'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Previous Visits'] >= lower_bound) & (df['Previous Visits'] <= upper_bound)]

# COLUMN: Conversion Rate

# identify the quartiles
q1, q3 = np.percentile(df['Conversion Rate'], [25, 75])

# calculate the interquartile range
iqr = q3 - q1

# calculate the lower and upper bounds
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)

# drop the outliers
traffic_clean = df[(df['Conversion Rate'] >= lower_bound) & (df['Conversion Rate'] <= upper_bound)]

# checking if outliershave been dropped successfully
print("Dropped outliers successfully!")
```
